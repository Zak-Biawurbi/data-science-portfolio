```python
import pandas as pd
```

```python
def grading_system(): #a fuction that takes the number of courses as an argument
    #Checking the number of courses the user takes
    no_courses = int(input("How many courses do you take: "))
    
    #Creating a dictionary to store the course and grade, numerical equivalence and grade points
    df ={
        'Course': [],
        'Grade': [],
        'Num_Equiv': [],
        'Grade Points': []
    }
    
    #Defining a variable to stare total credit hours
    total_credit_hours = 0
    
    
    for i in range(no_courses):
        
        #Checking the "nth" course in order to address the question
        if no_courses ==1:
            course = str(input("Kindly input the Course Name: "))

        elif i == no_courses-1:
            course = str(input("Kindly input the Course Name: "))

        elif i ==0:
            course = str(input("Kindly input the Course Name: "))

        else:
            course = str(input("Kindly input the Course Name: "))
            
        #Adiing course to the df
        df["Course"].append(course)

        #Using the score for the particular course
        score = float(input(f"What score did you get in {course}: "))       

        #Taking the credit hours per course
        credit_hours = float(input(f"What is the credit hours for {course}"))
        
        #Adding credit hours to total credit hours
        total_credit_hours += credit_hours
        
        #Making sure the user inputa a valid score before he can continue
        while True:
            
            if score<101 and score>=0:
                break
            
            else:
                score = float(input(f'Your {course} score should be between 0 and 100. Please enter an accurate score.'))


        #Checking the grade and adding it
        if score <101 and score >84:
            df['Grade'].append("A+")
            df['Num_Equiv'].append(4)
            
        elif score < 85 and score >79:
            df['Grade'].append("A")
            df['Num_Equiv'].append(3.5)
            
        elif score < 80 and score >74:
            df['Grade'].append("B+")
            df['Num_Equiv'].append(3)
            
        elif score < 75 and score >69:
            df['Grade'].append("B")
            df['Num_Equiv'].append(2.5)
            
        elif score < 70 and score >64:
            df['Grade'].append("C+")
            df['Num_Equiv'].append(2)
            
        elif score < 65 and score >59:
            df['Grade'].append("C")
            df['Num_Equiv'].append(1.5)
            
        elif score < 60 and score >54:
            df['Grade'].append("D+")
            df['Num_Equiv'].append(1)
            
        elif score < 55 and score >49:
            df['Grade'].append("D")
            df['Num_Equiv'].append(.5)
            
        elif score<50 and score >=0:
            df['Grade'].append("F")
            df['Num_Equiv'].append(0)
            
        #Calculating the grade point and adding it to the grade point list
        df['Grade Points'].append(df['Num_Equiv'][-1]*credit_hours)
        
    #Calculating GPA and CWA
    gpa = sum(df['Grade Points'])/total_credit_hours
    cwa = (gpa/4)*100
       
    #changing the dictionary to dataframe and storing it as 'output' for display
    output = pd.DataFrame(df)  
    print(f"Your 4-Point (GPA) Grade Poing Average is {gpa}")
    print(f"Your (CWA) Cumulative Weighted Average is {cwa}")
    return output
     
    
grading_system()

```
